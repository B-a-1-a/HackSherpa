```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Devpost Hackathons URL
DEVPOST_HACKATHONS_URL = "https://devpost.com/hackathons"

def get_hackathon_links():
    """Extracts all hackathon names and URLs from 'hackathons-container' by continuous scrolling."""
    print(f"üîç Fetching hackathon data from {DEVPOST_HACKATHONS_URL}")

    # Set up Selenium WebDriver
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Run in headless mode (no UI)
    options.add_argument("--disable-gpu")
    options.add_argument("--window-size=1920x1080")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    # Start WebDriver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)

    # Open the Devpost Hackathons page
    driver.get(DEVPOST_HACKATHONS_URL)

    # Wait for "hackathons-container" to load
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "hackathons-container"))
        )
    except:
        print("‚ùå Timeout: 'hackathons-container' did not load.")
        driver.quit()
        return pd.DataFrame(columns=["Hackathon Name", "URL"])

    hackathon_data = set()  # Using a set to avoid duplicates

    last_count = 0
    scroll_pause_time = 2  # Adjust if necessary
    max_scrolls = 20  # Limit the number of scrolls to prevent infinite loops

    for i in range(max_scrolls):
        # Extract hackathons from currently loaded portion
        hackathons_container = driver.find_element(By.CLASS_NAME, "hackathons-container")
        hackathon_tiles = hackathons_container.find_elements(By.CLASS_NAME, "hackathon-tile")

        for tile in hackathon_tiles:
            try:
                # Extract hackathon name
                name_element = tile.find_element(By.CSS_SELECTOR, ".main-content h3.mb-4")
                hackathon_name = name_element.text.strip()

                # Extract hackathon link
                link_element = tile.find_element(By.TAG_NAME, "a")
                hackathon_url = link_element.get_attribute("href")

                hackathon_data.add((hackathon_name, hackathon_url))

            except Exception as e:
                print(f"‚ö†Ô∏è Skipping an entry due to missing elements: {e}")

        # Stop if no new hackathons were added
        if len(hackathon_data) == last_count:
            print("‚úÖ No more new hackathons found. Stopping.")
            break
        last_count = len(hackathon_data)

        # Scroll down to load more hackathons
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause_time)  # Wait for new content to load

    # Close the browser
    driver.quit()

    # Convert to DataFrame
    global hackathons_df  # Make accessible outside function
    hackathons_df = pd.DataFrame(hackathon_data, columns=["Hackathon Name", "URL"])

    # Print and return the DataFrame
    print("\n‚úÖ Extracted Hackathons:")
    print(hackathons_df)

    return hackathons_df


# Run the function to populate the dataframe (for Jupyter Notebook)
hackathons_df = get_hackathon_links()
```

### Popular Hackathon Projects Scraper

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

# Base URL for pagination
BASE_URL = "https://devpost.com/software/popular?page={}"
MAX_PAGES = 50

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

def get_project_links():
    """Iterates through up to 20 pages and collects all unique project links."""
    project_links = set()  # Using a set to prevent duplicates
    
    for page in range(1, MAX_PAGES + 1): 
        url = BASE_URL.format(page)
        print(f"Scraping: {url}")
        response = requests.get(url, headers=headers)

        if response.status_code != 200:
            print("Error fetching page or no more pages available.")
            break

        soup = BeautifulSoup(response.text, 'html.parser')
        projects = soup.select("a.block-wrapper-link.fade.link-to-software")

        if not projects:
            print("No more projects found, stopping.")
            break

        # Extract project links and ensure uniqueness
        for project in projects:
            project_links.add(project["href"])

    return list(project_links)  # Convert set back to list for further processing

# Extract all unique project links (up to 20 pages)
all_project_links = get_project_links()

print(f"Total Unique Projects Found: {len(all_project_links)}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import hashlib
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

def categorize_project(technologies):
    """Assigns a project category based on the technologies used."""
    categories = {
        "AI": ["tensorflow", "pytorch", "openai", "machine learning", "deep learning"],
        "Web": ["react", "django", "flask", "html", "css", "javascript", "next.js"],
        "Mobile": ["flutter", "react native", "swift", "kotlin", "android", "ios"],
        "Blockchain": ["solidity", "ethereum", "web3", "polygon", "nft", "smart contract"],
        "IoT": ["raspberry pi", "arduino", "iot", "hardware"],
        "Game": ["unity", "unreal engine", "godot", "game development"]
    }

    technologies = technologies.lower()
    for category, keywords in categories.items():
        if any(keyword in technologies for keyword in keywords):
            return category
    return "Other"

def generate_project_id(project_name, project_url):
    """Generates a unique ID based on project name and URL using SHA256 hash."""
    return hashlib.sha256(f"{project_name}{project_url}".encode()).hexdigest()[:10]

def extract_project_details(project_url):
    """Extracts key project details from a Devpost project page."""
    project_data = {
        "id": "", "Project Name": "", "Category": "", "Built With": "", 
        "Group Members": 0, "Description": "", "Project URL": project_url
    }

    try:
        response = requests.get(project_url, headers=headers)
        if response.status_code != 200:
            print(f"Error fetching {project_url}, skipping.")
            return None
        
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract Project Name
        title_element = soup.select_one("#app-title")
        if title_element:
            project_data["Project Name"] = title_element.text.strip()
            project_data["id"] = generate_project_id(project_data["Project Name"], project_url)

        # Extract "Built With" technologies
        built_with_elements = soup.select("span.cp-tag")
        technologies_used = ", ".join([tech.text.strip() for tech in built_with_elements])
        project_data["Built With"] = technologies_used
        project_data["Category"] = categorize_project(technologies_used)

        # Count team members
        team_section = soup.select_one("#app-team ul")
        if team_section:
            project_data["Group Members"] = len(team_section.find_all("li"))

        # Extract Description
        description_element = soup.select_one("div#app-details-left p")
        if description_element:
            project_data["Description"] = description_element.text.strip()
    
    except Exception as e:
        print(f"Error processing {project_url}: {e}")
        return None

    return project_data

def scrape_all_projects(project_links):
    """Visits each project URL and extracts details."""
    all_projects = []

    for project_url in project_links:
        project_data = extract_project_details(project_url)
        if project_data:
            all_projects.append(project_data)

    # Convert to DataFrame
    projects_df = pd.DataFrame(all_projects)
    return projects_df

# Run the scraper
projects_df = scrape_all_projects(all_project_links[:20])

# Save to CSV
projects_df.to_csv("test_projects.csv", index=False)
```

```python
import os

# Check if existing CSV file exists
csv_filename = "test_projects.csv"
if os.path.exists(csv_filename):
    print("Loading existing project data from CSV...")
    existing_projects_df = pd.read_csv(csv_filename)
else:
    print("No existing CSV found. Creating a new dataset...")
    existing_projects_df = pd.DataFrame(columns=["id", "Project Name", "Category", "Built With", "Group Members", "Description", "Project URL"])

# Run the scraper and get new project data
new_projects_df = scrape_all_projects(all_project_links)

# Merge old and new data
combined_df = pd.concat([existing_projects_df, new_projects_df], ignore_index=True)

# Remove duplicates based on 'Project URL'
cleaned_df = combined_df.drop_duplicates(subset=["Project URL"], keep="first")

# Save to CSV
cleaned_df.to_csv(csv_filename, index=False)

print(f"Scraping complete! {len(cleaned_df)} unique projects saved to '{csv_filename}'.") 
```

```python
import os
import requests
import csv
from bs4 import BeautifulSoup
from groq import Groq
import pandas as pd
from dotenv import load_dotenv


load_dotenv()
# Set up Groq API client with your API key
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

# Function to extract image URL under the specified id
def get_image_url(soup):
    # Find the element with the id 'app_details'
    app_details = soup.find(id='app_details')
    
    # If the element exists, find the first image inside it
    if app_details:
        first_image = app_details.find('img')
        if first_image and 'src' in first_image.attrs:
            return first_image['src']
    
    # Return None if no image is found or no element with id='app_details'
    return None


# Function to scrape data from the page
def scrape_page(url):
    # Send HTTP request
    response = requests.get(url)
    
    # If request is successful
    if response.status_code == 200:
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Extract relevant data
        title = soup.find('h1', {'id': 'app-title'}).text.strip()
        description = soup.find('p', {'class': 'large'}).text.strip()
        
        # Extract image URL under 'app_details'
        image_url = get_image_url(soup)

        # Prepare the data
        post_data = {
            "title": title,
            "description": description,
            "image_url": image_url
        }

        return post_data
    else:
        print("Failed to retrieve the page.")
        return None

# Function to use Groq API for summary and categorization
def generate_summary(post_data):
    # Construct the prompt for the Groq API (chat completion)
    prompt = f"""
    Given the following information, generate a 100-word summary of the project description and categorize it into a primary and secondary category.
    
    Title: {post_data['title']}
    Description: {post_data['description']}
    
    Choose Categories From STRICTLY from the list below. Do Not Add Additional Categories or Duplicate Categories. Pick the primary category that is the MOST relevant to the project, 
    and the secondary category that is the NEXT MOST relevant.
    
    Categories:
    Mobile Apps
    Web Apps
    Desktop Apps
    Game Development
    Virtual Reality (VR)
    Augmented Reality (AR)
    AI/Artificial Intelligence
    Machine Learning (ML)
    Natural Language Processing (NLP)
    Robotics
    IoT (Internet of Things)
    Wearables
    Bio-Tech/Health-Tech
    FinTech (Financial Technology)
    EduTech (Educational Technology)
    Blockchain/Web3
    Cybersecurity
    Data Science/Analytics
    Cloud Computing
    Big Data
    Environmental Tech
    Digital Art
    Social Impact
    Hardware Development (Arduino/Raspberry Pi)
    Embedded Systems
    Quantum Computing
    3D Printing
    Chatbots
    Geospatial Data/GIS
    Supply Chain/Logistics
    LegalTech
    GovTech
    Marketing Tech
    Human-Computer Interaction
    Content Creation
    AgriTech
    Energy Tech
    Transportation Tech
    
    
    Return the output STRICTLY in the format: 
    
    Primary Category: <primary_category> @ Secondary Category: <secondary_category> @ Summary: <summary>
    """
    
    # Send request to Groq API for chat completion
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        model="llama-3.3-70b-versatile",  # Use an appropriate model
    )
    
    # Get the response from Groq API
    output = chat_completion.choices[0].message.content.strip()
    
    print(output)
    
    # Split the output into categories and summary
    output_parts = output.split("@")
    primary_category = output_parts[0].split(":")[1].strip()
    secondary_category = output_parts[1].split(":")[1].strip()
    summary = output_parts[2].split(":")[1].strip()
    
    # Add the categories and summary to the data
    post_data["primary_category"] = primary_category
    post_data["secondary_category"] = secondary_category
    post_data["summary"] = summary
    
    return post_data

# Function to save the data to a Supabase-compatible CSV file
def save_to_csv(post_data):
    # Define the column names for the CSV
    fieldnames = ['id', 'title', 'description', 'image_url', 'primary_category', 'secondary_category', 'summary']
    
    # Load existing CSV or create a new one
    file_name = 'posts_data.csv'
    try:
        df = pd.read_csv(file_name)
        next_id = df['id'].max() + 1
    except FileNotFoundError:
        next_id = 1
        df = pd.DataFrame(columns=fieldnames)
    
    # Prepare new row to be added
    new_row = {
        'id': next_id,
        'title': post_data['title'],
        'description': post_data['description'],
        'image_url': post_data['image_url'],
        'primary_category': post_data['primary_category'],
        'secondary_category': post_data['secondary_category'],
        'summary': post_data['summary']
    }
    
    # Create a new DataFrame for the new row and concatenate it with the existing DataFrame
    new_row_df = pd.DataFrame([new_row])
    df = pd.concat([df, new_row_df], ignore_index=True)
    
    # Save to CSV
    df.to_csv(file_name, index=False)
    
    print("Data saved successfully in CSV!")


# Main function to scrape the page, generate summary, and save data
def main():
    url = 'https://devpost.com/software/hacksat'  # Replace with the actual URL
    post_data = scrape_page(url)
    
    if post_data:
        post_data_with_summary = generate_summary(post_data)
        save_to_csv(post_data_with_summary)
    else:
        print("Failed to process the page.")

if __name__ == '__main__':
    main()
```
